```python
import csv
import os
import glob
import pandas as pd
import re
import nltk
import webbrowser
from nltk.corpus import stopwords
from collections import Counter 
from nltk import ngrams
import random
from nltk import word_tokenize
```

```python
os.chdir('C://Users//Ruben//Documents//Scriptie//test-scripts///metadata')
csv_filenames = glob.glob('*.csv')
```

```python
def Clean(string):
    strip_special_chars = re.compile("[^A-Za-z]")
    string = re.sub(strip_special_chars," ", string)
    string = string.split()
    string = [w.lower() for w in string]
    #string = [word for word in string if len(word) > 1]
    #string = [word for word in string if word not in stopwords.words("dutch")]
    return(string)
```

```python
dict_kwic = dict()

for csv in csv_filenames:
    df = pd.read_csv(csv, sep = "\t")
    
    for x in range(0,len(df)-1):
        if "buitenland" in df.ocr[x]:
            i = Clean(df.ocr[x])
            
            if "buitenland" in i:
                p = [aa for aa, elem in enumerate(i) if 'buitenland' in elem]
                p = int(p[0])

                s = p - 12
                e = p + 12
                if s < 0:
                    s = 0
                if e > (len(i) - 1):
                    e = (len(i) - 1)


                k = i[s:e]
                k = ' '.join(k)
                k = {df.date[x]:k}
                dict_kwic.update(k)
```

```python
kwic_df = pd.DataFrame(list(dict_kwic.keys()))
kwic_df['kwic'] = pd.DataFrame(list(dict_kwic.values()))
kwic_df
```

```python
with open('kwics.txt', 'w') as f:
    for item in list(kwic_df.kwic):
        f.write("%s\n" % item)
```
