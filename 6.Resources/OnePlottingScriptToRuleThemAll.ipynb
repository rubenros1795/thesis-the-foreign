```python
os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//PROP//output-data")

df = pd.read_csv('pmi_temporality2.csv')
df
```

```python
d = pd.DataFrame()

for w in ['tegenwoordig', 'heden', 'vandaag']:
    t = df[(df['w1'] == "buitenland") & (df['w2'] == w)]
    t = t[t['year'] > 1829]
    t = pd.DataFrame(list(t['pmi']))
    t.columns = [w]
    d = pd.concat([d, t], axis=1)
d['year'] = list(range(1830,1915))
d
```

```python
x_values = list(d['year'])
y1 = d['tegenwoordig']
y2 = d['heden']
y3 = d['vandaag']

rm1 = y1.rolling(window=5).mean()
rm2 = y2.rolling(window=5).mean()
rm3 = y3.rolling(window=5).mean()



#title = "Creativity Scores for 'Buitenlandsche'-vocabulary"
#color_point1 = "lightgrey"
#color_point2 = "linen"
#color_point3 = "azure"
#color_point3 = "lightblue"


plt.figure(figsize=(15,8))

mpl_fig = plt.scatter(x = x_values, y = y1, s = 8, alpha = 0.4)
plt.scatter(x = x_values, y = y2, s = 8, alpha = 0.4)
plt.scatter(x = x_values, y = y3, s = 8, alpha = 0.4)

plt.plot(x_values, rm1, label = 'afstand')
plt.plot(x_values, rm2, label = 'betrekking')
plt.plot(x_values, rm3, label = 'verbinding')
#lgd = plt.legend(d.columns[:-1], loc='upper center', bbox_to_anchor=(0.5, -0.06))
axes = plt.gca()
axes.set_ylim([1,2])
plt.xlabel('Years')
plt.ylabel('PMI Score')
os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//Visualizations//CH3")

plt.savefig("collocations_temporality.png", format = 'png', dpi = 800)
plt.show()
```

** Stacked Area Plot (for clusters) **

```python
os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//Visualizations//CH1//source-files")
```

```python
col = sns.color_palette("tab20b", 11)
```

```python
for i in [12]:
    fn = "buitenlandsche_clusters_" + str(i) + ".csv"
    df = pd.read_csv(fn)
    df = df.iloc[:,1:]
    
    x_lab = "Years"
    y_lab = "Absolute Frequency"
    title = "Cluster Distribution of 'Buitenlandsche'-vocabulary with " + str(i) + " Clusters"
    
    
    fig, ax = plt.subplots(facecolor='white',figsize=(24, 8))
    labels = list(df.columns[:-2])
    plt.stackplot(list(df.year), *[df[col] for col in labels], baseline="wiggle", colors = col)
    ax.set_facecolor("white")
    box = ax.get_position()
    #hfont = {'fontname':'Helvetica'}

    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
    lgd = ax.legend(labels, loc='upper center', bbox_to_anchor=(0.5, -0.06))
    
    for ticks in ax.xaxis.get_ticklines() + ax.yaxis.get_ticklines():
        ticks.set_color("white")
    for pos in ['top', 'bottom', 'right', 'left']:
        ax.spines[pos].set_edgecolor("white")
    
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=18)

    plt.title(title, fontsize=22)


    #plt.title('title',**hfont)
    #plt.xlabel('xlabel', **hfont)
    
    fn = "cluster-plot-poster-" + str(i) + ".png"
    os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//Visualizations//CH1")
    fig.savefig(fn, facecolor=fig.get_facecolor(), dpi=800,bbox_extra_artists=(lgd,), bbox_inches='tight', transparent = True)
    os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//Visualizations//CH1//source-files")
```
