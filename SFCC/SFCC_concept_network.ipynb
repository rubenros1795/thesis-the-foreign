```python
def GetNetwork(word, year, similarity_cutoff, freq_threshold):
    mfn = [m for m in list_mod if int(m[0:4]) == year]
    
    #model = KeyedVectors.load_word2vec_format(mfn, binary = True)
    model = KeyedVectors.load(mfn[0], mmap='r')
    
    
    ttp = tt[(tt.year >= year) & (tt.year < (year + 10))]
    ttp = ttp['tokens'].sum()
    
    dft = pd.DataFrame()
    
    for ms in list(dict(model.most_similar(word, topn=30)).keys()):

        first_list = list(dict(model.most_similar(ms, topn=30)).keys())

        first_df = pd.DataFrame()

        for term in first_list:
            tmp_df = pd.DataFrame([[key,value] for key,value in dict(model.most_similar(term, topn=30)).items()], columns = ['target', 'weight'])
            tmp_df['source'] = term
            first_df = first_df.append(tmp_df)

        dft = dft.append(first_df)
    
    dft['count'] = [model.wv.vocab[w].count for w in list(dft.target)]
    dft['count'] = dft['count'] / ttp
    
    dft = dft[dft['count'] > freq_threshold]
    dft = dft[dft.weight > similarity_cutoff]
    
    dgc = Counter(list(dft.target))
    dgc = [k for k,v in dict(dgc).items() if v >= 100]
    
    dft['degree'] = 1
    
    dft = dft[dft.target.isin(dgc)]
    
    return dft
```

```python
dft = pd.DataFrame()

for i in range(1835,1906,10):
    df = GetNetwork("onzekere_toestand", i, 0.3, 0.0000005)
    df['year'] = i
    dft = dft.append(df)
```
